# Country Populations

**Understanding basic Python functions/analysis through country populations**

```python
countries = ["USA", "India", "China", "Brazil", "Germany"]
populations = [331, 1400, 1440, 213, 83]
```

**Lists** in Python begin with 0

```python
countries[0]
```

Inputting a negative position in a list starts from the end of the list

```python
populations[-1]
```

A **for loop** runs until the range is reached, that is why a variable like 'i' can be used.
i runs through the countries and populations lists, going from i=0 to i=4

```python
for i in range(len(countries)):
    print(countries[i], "-", populations[i], "million")
```

**if, elif, and else** can be used as conditionals in a for loop, to designate what information is factored into
any given outcome

```python
for i in range(len(countries)):
    if populations[i] > 500:
        print(countries[i], "is very large")
```

```python
total = sum(populations)
average = total / len(populations)
print("Average Population: ", average)
```

```python
type(average)
```

A **dictionary** is like an advanced list, of sorts, following the "Key: Value" structure. This way,
every country(key) is assigned it's own population value. We can then further denote each key and value.
Ex: "USA": 331, can become country: population if we run a for loop through the dictionary items.

```python
country_data = {
    "USA": 331,
    "India": 1400,
    "China": 1440,
    "Brazil": 213,
    "Germany": 83
}
```

```python
for country, population in country_data.items():
    print(country, "has", population, "million people")
```

A **function** is a block of code that can be reused by calling it's name. In this case "classify_country" is being used to 
determine the size of a country by returning a size classification.

```python
def classify_country(population):
    if population > 1000:
        return "Huge"
    elif population > 300:
        return "Large"
    else:
        return "Medium/Small"
```

```python
for country, population in country_data.items():
    size = classify_country(population)
    print(country, "is classified as", size)
```

```python
country_data = {
    "USA": 331,
    "India": 1400,
    "China": 1440,
    "Brazil": 213,
    "Germany": 83,
    "Japan": 125
}
```

```python
def classify_country(population):
    if population > 1200:
        return "Mega"
    elif 800 < population < 1200:
        return "Huge"
    elif 300 < population < 800:
        return "Large"
    else:
        return "Small"
```

```python
for country, population in country_data.items():
    size = classify_country(population)
    print(country, "is classified as", size)
```
